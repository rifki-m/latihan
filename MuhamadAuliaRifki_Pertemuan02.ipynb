```python
def getKali(numbers):
    result = 1
    for num in numbers:
        result *= num
    return result
myNumbers = [2, 3, 4, 5]
product = getKali(myNumbers)
print("Hasil perkalian:", product)
```

```python
def getBagi(numbers):

    if len(numbers) == 0:
        return None

    result = numbers[0]
    for num in numbers[1:]:
        result /= num
    return result

myNumbers = [10, 5, 2]
division = getBagi(myNumbers)
print("Hasil pembagian:", division)
```

```python
def getBagiHimpunan(himpunan1, himpunan2):

    if len(himpunan1) != len(himpunan2):
        return None  # Jika kedua himpunan tidak memiliki panjang yang sama, kembalikan None

    hasil = []
    for i in range(len(himpunan1)):
        hasil.append(himpunan1[i] / himpunan2[i])
    return hasil
```

```python
himpunan_a = [10, 20, 30]
himpunan_b = [2, 5, 6]

hasil_bagi = getBagiHimpunan(himpunan_a, himpunan_b)
print("Hasil pembagian:", hasil_bagi)
```

```python
def getKurangHimpunan(himpunan1, himpunan2):

    if len(himpunan1) != len(himpunan2):
        return None  # Jika kedua himpunan tidak memiliki panjang yang sama, kembalikan None

    hasil = []
    for i in range(len(himpunan1)):
        hasil.append(himpunan1[i] - himpunan2[i])
    return hasil
```

```python
himpunan_a = [10, 20, 30]
himpunan_b = [2, 5, 6]

hasil_kurang = getKurangHimpunan(himpunan_a, himpunan_b)
print("Hasil pengurangan:", hasil_kurang)
```
