```python
bilangan1 = 3
bilangan2 = 4
kpk = max(bilangan1, bilangan2)

while True:
    if kpk % bilangan1 == 0 and kpk % bilangan2 == 0:
        break
    kpk += max(bilangan1, bilangan2)

print("KPK dari", bilangan1, "dan", bilangan2, "adalah", kpk)
```
